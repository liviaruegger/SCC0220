```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## **TESTE 1 (valores entre 0 e 10000)**

```python
teste1_counting_original = pd.read_csv('teste1/counting.csv')
teste1_bucket_original   = pd.read_csv('teste1/bucket.csv')
teste1_radix_original    = pd.read_csv('teste1/radix.csv')
teste1_quick_original    = pd.read_csv('teste1/quick.csv')
teste1_heap_original     = pd.read_csv('teste1/heap.csv')
teste1_merge_original    = pd.read_csv('teste1/merge.csv')

counting1 = teste1_counting_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)
bucket1   = teste1_bucket_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)
radix1    = teste1_radix_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)
quick1    = teste1_quick_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)
heap1     = teste1_heap_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)
merge1    = teste1_merge_original[['vector_size', 'execution_time']].groupby('vector_size').agg(np.mean)

plt.figure(figsize=[10, 10])
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo de execuo mdio [s]')

plt.plot(counting1['execution_time'], 'o-', label='Counting Sort')
plt.plot(bucket1['execution_time'], 'o-', label='Bucket Sort')
plt.plot(radix1['execution_time'], 'o-', label='Radix Sort')
plt.plot(quick1['execution_time'], 'o-', label='Quick Sort')
plt.plot(heap1['execution_time'], 'o-', label='Heap Sort')
plt.plot(merge1['execution_time'], 'o-', label='Merge Sort')

plt.legend()
plt.savefig('teste1/0101.pdf')
```
